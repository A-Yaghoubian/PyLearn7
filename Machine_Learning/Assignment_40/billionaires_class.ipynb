```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('Billionaire.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
data.isnull().sum()
```

```python
data['NetWorth'] = data['NetWorth'].str.lstrip('$')
data['NetWorth'] = data['NetWorth'].str.rstrip('B')
data['NetWorth'] = data['NetWorth'].astype(float)
data['NetWorth']
```

```python
df = data.sort_values(by=['NetWorth'], ascending=False).head(10).reset_index()
df
```

```python
plt.figure(figsize=(24, 8))
plt.bar(df['Name'], df['NetWorth'])
plt.title('Top 10 in 2021')
plt.show()
```

```python
df = data['Source'].value_counts().head(10).reset_index()
df
```

```python
plt.pie(df['Source'], labels=df['index'])
plt.title('Top 10 Domains to become a Billionares')
plt.show()
```

```python
df = data['Industry'].value_counts().reset_index()
df
```

```python
plt.figure(figsize=(24, 10))
plt.pie(df['Industry'], labels=df['index'])
plt.title('Top 10 Industries to become a Billionares')
plt.show()
```

```python
df = data['Country'].value_counts().head(10).reset_index()
df
```

```python
plt.pie(df['Country'], labels=df['index'])
plt.title("Top 10 Countries to become a billionares")
plt.show()
```
